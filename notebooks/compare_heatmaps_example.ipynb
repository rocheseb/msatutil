Here we want to compare two L2 versions of the same granules but processed with different inputs
In the **O2_5x1** folder are L2 products from the O2 window using L1B spectra that include a straylight correction with a possible implementation issue at this time (2022-07-11)
In the **O2_5x1_nostralight** folder the same L1B granules are processed without the stray light correction.

The **compare_heatmaps** package can be used to quicly compare two limited sets of granules by displaying a side-by-side heatmap for a given field using the same color scale and by showing histograms of both datasets

### Load data into msat_collection objects

```python
l2 = get_msat('O2_5x1')
l2_nostray = get_msat('O2_5x1_nostray')
```

### Use compare_heatmaps

```python
ion()
fig,ax,x1,x2 = compare_heatmaps(l2,l2_nostray,['L2','L2 no straylight'],'dp',vminmax=[-25,25],ylim=[0,256],hist_xlim=[-30,50],exp_fmt=False)
```

The above plots suggest the straylight correction might have an implementation issue as the L1B files that include it present as ~20 hPa bias in the retrieved minus prior surface pressure, which is expected to have a distribution centered close to 0
